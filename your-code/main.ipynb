## Bonus: For each of the data sets in this lab, try training with some of the other models you have learned about, recalculate the evaluation metrics, and compare to determine which models perform best on each data set.

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

def evaluate_all_classifiers(X_train, X_test, y_train, y_test):
    """
    Initializes, trains, and evaluates a suite of standard classification models.
    """
    
    # Initialize models inside the function
    classifiers = {
        "Logistic Regression": LogisticRegression(max_iter=500),
        "KNN": KNeighborsClassifier(),
        "Decision Tree": DecisionTreeClassifier(),
        "Random Forest": RandomForestClassifier(),
        "SVM": SVC(probability=True),
        "XGBoost": xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
    }
    
    performance_summary = {}

    for name, model in classifiers.items():
        # Training the model
        # .values.ravel() converts the target DataFrame column into a 1D array
        model.fit(X_train, y_train.values.ravel())
        
        # Predicting outcomes
        predictions = model.predict(X_test)
        
        # Metrics calculation
        accuracy = accuracy_score(y_test, predictions)
        performance_summary[name] = accuracy

        print("=" * 50)
        print(f"MODEL: {name}")
        print(f"Global Accuracy: {accuracy:.4f}")
        print("=" * 50)
        
        print("\nConfusion Matrix:")
        print(confusion_matrix(y_test, predictions))
        
        print("\nDetailed Classification Report:")
        print(classification_report(y_test, predictions))
        print("\n")
    
    # Return the dictionary sorted by accuracy (highest first)
    return dict(sorted(performance_summary.items(), key=lambda item: item[1], reverse=True))

# --- Execution ---

# Assuming X_train_scaled, X_test_scaled, y_train, and y_test are already defined
best_performing_models = evaluate_all_classifiers(X_train, X_test, y_train, y_test)

# Print a quick ranking
print("--- MODEL RANKING ---")
for rank, (name, score) in enumerate(best_performing_models.items(), 1):
    print(f"{rank}. {name}: {score:.4f}")
```
